```python
# Problem Statement: Design a function to manage a smart parking lot system.
      # 1. Accept vehicle entry and exit logs
      # 2. Calculate total parked vehicles
      # 3. Identify peak parking usage
      # 4. Alert if parking exceeds capacity

print("----- 1. Smart Parking Lot Management System -----")

def smart_parking(capacity, logs):
    parked = 0  # Store current parked vehicles
    peak = 0    # Store peak usage

    for log in logs:  # Loop through vehicle logs
        if log == "IN":
            parked += 1  # Increase count for entry
        elif log == "OUT" and parked > 0:
            parked -= 1  # Decrease count for exit

        if parked > peak:
            peak = parked  # Update peak usage

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)

    if parked > capacity:
        print("Parking Status: Full - Alert!")
    else:
        print("Parking Status: Available")


# Sample Input
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Problem Statement: Create a function to estimate delivery time based on:
      # ● Distance (km)
      # ● Traffic level
      # ● Weather condition

print("----- 2. Online Food Delivery Time Estimator -----")

def delivery_time(distance, traffic, weather):
    eta = distance * 5  # Base time calculation

    if traffic == "High":
        eta += 15  # Add delay for high traffic
    elif traffic == "Medium":
        eta += 10  # Add delay for medium traffic

    if weather == "Rainy":
        eta += 10  # Add delay for rain
    elif weather == "Stormy":
        eta += 20  # Add delay for storm

    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
delivery_time(8, "High", "Rainy")
```

```python
# Problem Statement: Build a function that:
      # 1. Calculates occupancy percentage
      # 2. Determines if show is Housefull
      # 3. Suggests opening additional shows

print("----- 3. Movie Theatre Seat Occupancy Analyzer -----")

def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)  # Count booked seats
    occupancy = (booked / total_seats) * 100  # Calculate occupancy

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
seat_occupancy(200, [1]*150)
```

```python
# Problem Statement: Create a function to classify server load based on CPU usage readings.
      # ● Average CPU < 50% → Normal
      # ● 50%–80% → Warning
      # ● 80% → Critical

print("----- 4. Cloud Server Load Classification System -----")

def server_load(cpu_readings):
    total = 0  # Store total CPU

    for value in cpu_readings:
        total += value  # Add each CPU reading

    avg = total / len(cpu_readings)  # Calculate average

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
server_load([45, 60, 70, 85, 90])
```

```python
# Problem Statement: Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.

print("----- 5. Smart Classroom Resource Usage Monitor -----")

def resource_monitor(resources):
    overused = []  # Store overused resources

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)  # Add resource to list

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# Problem Statement: Create a function that manages event registrations by:
      # 1. Tracks registrations
      # 2. Prevents overbooking
      # 3. Triggers waitlist mode

print("----- 6. Online Event Registration Capacity Controller -----")

def event_registration(capacity, registrations):
    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        waitlist = registrations - capacity  # Calculate extra users
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")


# Sample Input
event_registration(100, 105)
```
