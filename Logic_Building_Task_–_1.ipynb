```python
# Problem Statement: Given a username and password, check whether login is successful.

print("----- 1. User Login Check -----")

# Default admin credentials
admin_username = "admin"
admin_password = "1234"

# Input from the user
username = input("Enter username: ")
password = input("Enter password: ")

# Login validation
if username == admin_username and password == admin_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Problem Statement: Analyze student results.

print("----- 2. Pass / Fail Analyzer -----")

marks = [45, 78, 90, 33, 60]

# Counters to track number of passed and failed students
pass_count = 0
fail_count = 0

# From each student's marks classify as pass or fail
for student_marks in marks:
    if student_marks >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total No Of Students:",len(marks))
print()

# Display Each Student With Their Marks.
for i in range(1, len(marks) +1):
 print("Student No", i, "marks :",marks[i-1])
print()

print("Total No Of Passed Students:", pass_count)
print("Total No Of Failed Students:", fail_count)
print("Out Of 5 Students Pass_Count is :", pass_count, "And Fail_Count is :", fail_count)
```

```python
# Problem Statement: Clean and standardize user names.

print("----- 3. Simple Data Cleaner -----")

names = [" Alice ", "bob", " CHARLIE "]

# Create an empty list to store cleaned names.
cleaned_names = []

# Loop through names to remove spaces and convert to lowercase
for name in names:
    clean_name = name.strip().lower()
    cleaned_names.append(clean_name)

print("Cleaned Names:", cleaned_names)
```

```python
 # Problem Statement: Analyze message sizes.

print("----- 4. Message Length Analyzer -----")

messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message and display its content along with its length.
for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    # Check if the message length is 10 or more characters and flag it if true
    if length >= 10:
        print(f" '{message}' ", "-> This message is longer than 10 or more characters")
    print()
```

```python
# Problem Statement: Detect error messages from system logs.

print("----- 5. Error Message Detector -----")

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counter to track number of error_count
error_count = 0

# Loop through logs and count the number of error messages.
for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR messages:", error_count)
```
