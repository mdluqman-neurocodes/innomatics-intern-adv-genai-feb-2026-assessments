```python
# Problem Statement: Create a function to validate a mobile recharge request.

print("----- 1. Mobile Recharge Validation System -----")

# Function to validate recharge amount and selected plan
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]  # allowed recharge plans
    # Return True only if amount >= 50 AND plan is valid
    return amount >= 50 and plan in valid_plans


# Keep asking user until valid input is provided
while True:
    amount = float(input("Enter recharge amount: "))
    plan = int(input("Enter plan (199, 299, 399, 599): "))

    # Call validation function
    if validate_recharge(amount, plan):
        print("Recharge Successful ✅")
        break  # exit loop if valid
    else:
        print("Invalid recharge details. Please try again ❌")
```

```python
# Problem Statement: Given a dictionary of products and stock quantities: ● Print Reorder Alert if stock < 15 ● Print Stock OK otherwise

print("----- 2. Inventory Reorder Alert System -----")

# Function to check stock levels of products
def check_inventory(products):
    # Loop through each product and its stock quantity
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert ⚠️")  # low stock warning
        else:
            print(f"{product}: Stock OK ✅")  # sufficient stock


# Sample product dictionary
products = {
    "Rice": 10,
    "Oil": 20,
    "Sugar": 5,
    "Flour": 18
}

check_inventory(products)
```

```python
# Problem Statement: Create a function that: ● Accepts a list of student marks ● Calculates average marks ● Prints Pass if average ≥ 50 else Fail

print("----- 3. Student Result Processing System -----")

# Function to calculate average and determine result
def process_result(marks):
    total = 0

    # Calculate total marks using loop
    for mark in marks:
        total += mark

    # Compute average
    average = total / len(marks)

    # Return Pass if average >= 50 else Fail
    if average >= 50:
        return "Pass ✅"
    else:
        return "Fail ❌"


# Example marks list
marks = [60, 70, 45, 80]

print("Result:", process_result(marks))
```

```python
# Problem Statement: Create a cab fare calculator:
   # ● Base fare = ₹50
   # ● ₹12 per km
   # ● Peak hour → 25% extra
   # ● Allow retry using while loop

print("----- 4. Cab Fare Estimator with Retry Option -----")

# Function to calculate total cab fare
def calculate_fare(km, peak):
    # Base fare + distance charge
    fare = 50 + (12 * km)

    # Add 25% extra during peak hours
    if peak:
        fare += 0.25 * fare

    return fare


# Allow user to retry fare calculation
while True:
    km = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()

    peak = True if peak_input == "yes" else False

    total_fare = calculate_fare(km, peak)
    print("Total Fare: ₹", round(total_fare, 2))

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break  # exit loop
```

```python
# Problem Statement: Create a function that:
    # ● Accepts attendance list ("P", "A")
    # ● Calculates attendance percentage
    # ● Returns Eligible if ≥ 75% else Not Eligible

print("----- 5. Employee Attendance Eligibility -----")

# Function to calculate attendance percentage
def check_attendance(attendance_list):
    present_days = 0

    # Count number of present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / len(attendance_list)) * 100

    # Return eligibility based on 75% rule
    if percentage >= 75:
        return "Eligible ✅"
    else:
        return "Not Eligible ❌"


# Example attendance record
attendance = ["P", "P", "A", "P", "P"]

print("Attendance Status:", check_attendance(attendance))
```

```python
# Create a function to check password strength.
    # Rules:
    # ● Minimum length = 8
    # ● Must contain at least one digit
    # ● Must contain at least one special character (@#$)

print("----- 6. Password Strength Checker -----")

# Function to check password strength based on given rules
def check_password(password):

    # Rule 1: Minimum length check
    if len(password) < 8:
        return "Weak Password ❌"

    has_digit = False
    has_special = False
    special_chars = "@#$"  # allowed special characters

    # Loop through each character in password
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    # Password is strong only if both conditions are met
    if has_digit and has_special:
        return "Strong Password ✅"
    else:
        return "Weak Password ❌"


pwd = input("Enter password: ")
print(check_password(pwd))
```
