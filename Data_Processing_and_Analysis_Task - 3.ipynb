```python
# Problem Statement 1: Employee Performance Bonus Eligibility.

print("----- 1. Employee Performance Bonus Eligibility -----")

# Employee names and their performance scores.
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find employees with highest score
top_performers = []

for name, score in employees.items():
    if score == max_score:
        top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis

print("----- 2. Search Query Keyword Analysis -----")

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Split words
words = query.split()

# Count frequency
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Display words searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

```python
# Problem Statement 3: Sensor Data Validation

print("----- 3. Sensor Data Validation -----")

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store valid (even) readings here
valid_readings = []

# Loop through readings with their index (hour)
for index, value in enumerate(sensor_readings):
    # Keep only even readings
    if value % 2 == 0:
        # Save (hour, reading) as a tuple
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem Statement 4: Email Domain Usage Analysis

print("----- 4. Email Domain Usage Analysis -----")

# List of email addresses
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Extract domain from each email and count occurrences
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Total number of emails
total = len(emails)

# Calculate and print percentage for each domain
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem Statement 5: Sales Spike Detection

print("----- 5. Sales Spike Detection -----")

# Daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average = sum(sales) / len(sales)

# Set 30% above average limit
threshold = average * 1.30

# Loop through sales with index
for index, value in enumerate(sales):
    if value > threshold:  # Check if sale is a spike
        print(f"Day {index + 1}: {value}")  # Print day number and sale
```

```python
# Problem Statement 6: Duplicate User ID Detection

print("----- 6. Duplicate User ID Detection -----")

# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store count of each user ID
frequency = {}

# Loop through each user ID
for user in user_ids:
    frequency[user] = frequency.get(user, 0) + 1  # Increase count (start from 0 if new)

# Loop through dictionary items
for user, count in frequency.items():
    if count > 1:  # Check if user ID appears more than once
        print(f"{user} → {count} times")  # Print duplicate user and count
```
