```python
# Problem Statement: Given a sentence, identify all unique words and display.

print("----- 1. Unique Words in a Sentence -----")

sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Create empty list to store unique words
unique_words = []

for word in words:
    # If the word is not already in the list, add it
    if word not in unique_words:
        unique_words.append(word)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem Statement: Store employee names and their salaries, and determine which employee has the highest salary.

print("----- 2. Highest Salary from Employee Data -----")

# Create a dictionary to store employee names and their salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Create variables to store the highest salary and employee name
highest_name = ""
highest_salary = 0

for name in employees:
    salary = employees[name]

    # Compare current salary with highest_salary if current salary is greater, update highest_salary and highest_name
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name

print("Highest Salary:", highest_name, "-", highest_salary)
```

```python
# Problem Statement: Find the maximum and minimum values from a list of numbers.

print("----- 3. Find Maximum and Minimum Values -----")

numbers = [45, 22, 89, 10, 66]

# Assume first number as max and min
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    # Update maximum if current number is bigger
    if num > maximum:
        maximum = num
    # Update minimum if current number is smaller
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Problem Statement: Count how many products have a price greater than 1000.

print("----- 4. Count Products Above a Price Threshold -----")

prices = [450, 1200, 899, 1500, 300]

# Variable to count products above 1000
count = 0

for price in prices:
    # If price is greater than 1000, increase count
    if price > 1000:
        count = count + 1

print("Products above 1000:", count)
```

```python
# Problem Statement: Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

print("----- 5. Calculate Attendance Percentage -----")

# List showing attendance (P = Present, A = Absent)
attendance = ["P", "P", "A", "P", "P"]

# Variable to count present days
present_days = 0

for day in attendance:
    # If the day is Present, increase count
    if day == "P":
        present_days = present_days + 1

total_days = len(attendance)

# Calculate attendance percentage
percentage = (present_days / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem Statement: You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.

print("----- 6. Remove Duplicate Phone Numbers -----")

phone_numbers = [9876543210, 9123456789, 9876543210]

# Empty list to store unique numbers
unique_numbers = []

for number in phone_numbers:
    # If number is not already in unique list, add it
    if number not in unique_numbers:
        unique_numbers.append(number)

print("Unique phone numbers:", unique_numbers)
```

```python
# Problem Statement: Count how many times each character appears in a given string.

print("----- 7. Count Character Frequency -----")

# String to count characters from
text = "python"

# Empty dictionary to store character counts
char_count = {}

for char in text:
    # If character already exists in dictionary, increase its count
    if char in char_count:
        char_count[char] = char_count[char] + 1
    # If character appears for the first time, set count to 1
    else:
        char_count[char] = 1

print(char_count)
```

```python
# Problem Statement: Check whether a specific key exists in a dictionary.

print("----- 9. Check if a Key Exists in a Dictionary -----")

# Dictionary storing employee names and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Take employee name as input from user
name = input("Enter employee name: ")

# Check if the entered name is a key in the dictionary
if name in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem Statement: Calculate the average marks of a student from a list of marks.

print("----- 10. Calculate Average Marks -----")

# List of student marks
marks = [70, 75, 80, 65]

# Variable to store total marks
total = 0

for mark in marks:
    total = total + mark

# Calculate average (total marks divided by number of subjects)
average = total / len(marks)

print("Average Marks:", average)
```
